```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Arial'
# plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams['font.size'] = 18
# plt.rcParams['axes.linewidth'] = 2
plt.rcParams['figure.figsize'] = (10.0, 7.0)
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
sns.set_palette("Dark2")
```

```python
pubmed = pd.read_excel("../../publication_by_year_20201231.xlsx", sheet_name='pubmed')
# dimensions = pd.read_excel("../../publication_by_year_20201027.xlsx", sheet_name='dimensions')
```

```python
pubmed = pubmed[pubmed['Type'] != 'Image-omics']
```

```python
pubmed = pubmed[pubmed['num_publications'] > 0]
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(x='Year',
                  y='num_publications',
                  hue='Type',
                  hue_order=[
                      'Genomics', 'Epigenomics', 'Transcriptomics',
                      'Proteomics', 'Multi-omics'
                  ],
#                   style="Type",
                  data=pubmed[pubmed['Year'] >= 1995])
ax.set_yscale('log')
ax.set_ylabel("Number of publications (in log)")
plt.legend(bbox_to_anchor=(1.02, 1),
           loc=2,
           borderaxespad=0.,
           prop={'size': 18})
plt.title("Number of omic publications by year")
plt.tight_layout()
plt.savefig("../plots/Figure1.pdf", dpi=500)
```
